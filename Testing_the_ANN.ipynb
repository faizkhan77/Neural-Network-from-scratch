## Importing the Neural Network module and Testing it

```python
import NeuralNetwork
import pandas as pd
import numpy as np
```

I have already Preprocessed the data previously to save us time, if you want to check how i did the Data Cleaning, Preprocessing and deep EDA then check this [Notebook]("https://www.kaggle.com/code/faizkhan7/parkinsons-disease-analysis-and-prediction#6.-Data-Preprocessing") out 

```python
# Loading the data
df = pd.read_csv("datas//preprocessed_parkinson.csv")
```

```python
df
```

```python
# X and y split
X = df.drop(["Diagnosis"], axis=1)
y = df["Diagnosis"]
```

```python
X
```

```python
y
```

```python
# Train and Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
len(X_train), len(X_test)
```

```python
y_train.value_counts()
```

I will just quickly handle the Imbalance using SMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy="minority")
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
y_train_smote.value_counts()
```

## Training the Neural Network

```python
myNN = NeuralNetwork.NeuralNetwork()
```

```python
# Using Batch Gradient Descent Algorithm
myNN.fit(X_train, y_train, optimizer="batch", epoch=1000, learning_rate=0.1)
print(f"The Accuracy Score is: {myNN.score(X_test, y_test)}")
```

81% Score is pretty impressive for me as its my first attempt of building this Simple Neural Network from scratch

```python
# Using Mini Batch Gradient Descent Algorithm
myNN.fit(X_train, y_train, optimizer="mini", size=20, epoch=1000, learning_rate=0.01)
print(f"The Accuracy Score is: {myNN.score(X_test, y_test)}")
```

Well it performed the same and a gave a good accuracy of 81%

```python
# Using Stochastic Gradient Descent Algorithm
myNN.fit(X_train, y_train, optimizer="sgd", epoch=1000, learning_rate=0.01)
print(f"The Accuracy Score is: {myNN.score(X_test, y_test)}")
```

Since SGD only uses 1 random sample its understanable

Now lets do Training again and do Predictions

We can use `verbose=False` to off all those log messages

```python
myNN.fit(X_train, y_train, optimizer="batch", epoch=1000, learning_rate=0.1, verbose=False)
predict = myNN.predict(X_test)
predict = [1 if i > 0.5 else 0 for i in predict]
```

Now lets Evaluate

```python
myNN.score(X_test, y_test)
```

Lets also check the ROC_AUC score, Classification Report and Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
roc_auc_score(y_test, predict)
```

```python
print(classification_report(y_test, predict))
```

The report is prety neat!

```python
import matplotlib.pyplot as plt
import seaborn as sns
cm = confusion_matrix(y_test, predict)
plt.figure(figsize=(10, 7))
plt.title("Truth vs Predicted")
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted Values")
plt.ylabel("Truth Values")
```

```python
print(f"Total num of Predictions: {len(predict)}")
print(f"Total Wrong Predictions: {np.sum(predict != y_test)}")
```
